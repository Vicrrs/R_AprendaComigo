# Distribuição de Frequência

O primeiro passo em um trabalho de análise é o conhecimento do comportamento das variáveis envolvidas no estudo. Utilizando técnicas estatśticas como as análises das DISTRIBUIÇÕES DE FREQUÊNCIA e HISTOGRAMAS podemos avaliar melhor a forma como os fenômenos em estudo se distribuem

## Distribuição de Frequência para variáveis qualitativas

Método 1:

    * table
    * prop.table
    * cbind
    * row+colnames

```r
table(dados$Sexo)
```

```r
prop.table(table(dados$Sexo))*100
```

```r
dist_freq_qualitativas <- cbind(freq = table(dados$Sexo), percent=prop.table(table(dados$Sexo))*100)
dist_freq_qualitativas
```

```r
colnames(dist_freq_qualitativas) <- c('Frequência', 'Porcentagem (%)')
dist_freq_qualitativas
```

```r
rownames(dist_freq_qualitativas) <- c('Masculino', 'Feminino')
dist_freq_qualitativas
```

Método 2:

    * tapply

```r
frequencia <- table(dados$Sexo, dados$Cor)
frequencia
```

```r
rownames(frequencia) <- c('Masculino', 'Feminino')
frequencia
```

```r
colnames(frequencia) <- c('Indígena', 'Branca', 'Preta', 'Amarela', 'Parda')
frequencia
```

```r
frequencia <- cbind(frequencia)
frequencia
```

```r
percentual <- prop.table(frequencia) * 100
percentual
```

```r
# Cruzando uma terceira variavel -> Renda
medias <- tapply(dados$Renda, list(dados$Sexo, dados$Cor), mean)

rownames(medias) <- c('Masculino', 'Feminino')
colnames(medias) <- c('Indígena', 'Branca', 'Preta', 'Amarela', 'Parda')
```

```r
medias
```

## Distribuição de Frequência para variáveis quantitativas (classes personalizadas)

Documentação

    * cut
    * order
    * row.names

```r
min(dados$Renda)
```

```r
max(dados$Renda)
```

```r
classes <- c(0, 1576, 3152, 7880, 15760, 200000)

labels <- c('E', 'D', 'C', 'B', 'A')
```

Criando Tabela de Frequências

```r
cut(
    x = dados$Renda,
    breaks = classes,
    labels = labels,
    include.lowest = TRUE
)
```

```r
frequencia <- table(
    cut(
    x = dados$Renda,
    breaks = classes,
    labels = labels,
    include.lowest = TRUE
    )
)

frequencia
```

```r
percentual <- prop.table(frequencia) * 100
percentual
```

```r
dist_freq_quantitativas_personalizadas <- cbind('Frequencia' = frequencia, 'Porcentagem(%)' = percentual)

dist_freq_quantitativas_personalizadas
```

```r
dist_freq_quantitativas_personalizadas[
    order(row.names(dist_freq_quantitativas_personalizadas)),
    
]
```

## Distribuição de Frequência para variáveis quantitativas (classes de amplitude fixa)

Documentacao

    * nrow
    * log
    * round


1- Definindo o número de classes no histograma (https://geokrigagem.com.br/numero_classes_histograma_regra_sturges/)

Regra de Sturges

k = 1 + (10/3)* log(n)

```r
n <- nrow(dados)
n
```

```r
k <- 1 + (10/3) * log10(n)
k
```

```r
k <- round(k) # funcao de arredondamento
k
```

Criar a tabela de frequências

```r
labels <- c(
    '      0.00 |—|  11,764.70', 
    ' 11,764.70  —|  23,529.40', 
    ' 23,529.40  —|  35,294.10', 
    ' 35,294.10  —|  47,058.80', 
    ' 47,058.80  —|  58,823.50', 
    ' 58,823.50  —|  70,588.20', 
    ' 70,588.20  —|  82,352.90', 
    ' 82,352.90  —|  94,117.60', 
    ' 94,117.60  —| 105,882.00', 
    '105,882.00  —| 117,647.00', 
    '117,647.00  —| 129,412.00', 
    '129,412.00  —| 141,176.00', 
    '141,176.00  —| 152,941.00', 
    '152,941.00  —| 164,706.00', 
    '164,706.00  —| 176,471.00', 
    '176,471.00  —| 188,235.00', 
    '188,235.00  —| 200,000.00'
)
```

```r
frequencia <- table(
    cut(
        x = dados$Renda,
        breaks = k,
        labels = labels,
        include.lowest = TRUE
    )
)
frequencia
```

```r
percentual <- prop.table(frequencia) * 100
percentual
```

```r
dist_frequencia_quantitativa_amplitude_fixa <- cbind('Frequencia' = frequencia, 'Porcentagem(%)' = percentual)
dist_frequencia_quantitativa_amplitude_fixa
```

## Histograma

O HISTOGRAMA é a representação gráfica de uma distribuição de frequências. É um gráfico formado por um conjunto de retângulos colocados lado a lado, onde a área de cada retângulo é proporcional a frequência da classe que ele representa.


Documentação:

    * package.graphics
    * hist
    * package.ggplot2
    * data.frame

```r
# Configuracao para aumentar ou diminuir o grafico
options(repr.plot.width = 12, repr.plot.height = 8)
```

```r
hist(dados$Altura)
```

```r
hist(
    x = dados$Altura,
    breaks = 'Sturges',
    col = 'lightblue',
    main = 'Histograma das Alturas',
    xlab = 'Altura',
    ylab = 'Frequência'
)
```

```r
# Instalando biblioteca específica
install.packages("ggplot2")

# Importando
library(ggplot2)
```

```r
ggplot(dados, aes(x = Altura)) + 
    geom_histogram(binwidth = 0.02, color = "black", alpha = 0.9) + 
    ylab("Frequência") + 
    xlab("Alturas") + 
    ggtitle('Histograma das Alturas') +
    theme(
        plot.title = element_text(size = 14, hjust = 0.5),
        axis.title.y = element_text(size = 12, vjust = +0.2),
        axis.title.x = element_text(size = 12, vjust = -0.2),
        axis.text.y = element_text(size = 10),
        axis.text.x = element_text(size = 10)
    )
```

```r
formatos <- theme(
        plot.title = element_text(size = 14, hjust = 0.5),
        axis.title.y = element_text(size = 12, vjust = +0.2),
        axis.title.x = element_text(size = 12, vjust = -0.2),
        axis.text.y = element_text(size = 10),
        axis.text.x = element_text(size = 10)
    )
```

```r
ggplot(dados, aes(x = Altura, y = ..density..)) + 
    geom_histogram(binwidth = 0.02, color = "black", alpha = 0.9) + 
    geom_density(color = 'green') +
    ylab("Frequência") + 
    xlab("Alturas") + 
    ggtitle('Histograma das Alturas') +
    formatos
```

```r
bar_chart <- data.frame(dist_freq_quantitativas_personalizadas)
bar_chart
```
