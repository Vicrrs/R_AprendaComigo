# Estruturas de Controle

```r
# If-Else basico
if (5 > 3) {
  print("5 é maior que 3!")
} else {
  print("5 não é maior que 3!")
}
```

```r
# If-Else aninhado
num <- 10

if (num > 0) {
  print("Número positivo!")
} else if (num < 0) {
  print("Número negativo!")
} else {
  print("Zero!")
}
```

```r
# For Loop básico
for (i in 1:5) {
  print(i)
}
```

```r
# For loop com vetor
frutas <- c("maçã", "banana", "cereja")

for (fruta in frutas) {
  print(fruta)
}
```

```r
# While Loop básico
contador <- 1

while (contador <= 5) {
  print(contador)
  contador <- contador + 1
}
```

```r
# Repeat loop
x <- 1

repeat {
  print(x)
  x <- x + 1
  if (x == 11) {
    break
  }
}
```

```r
# Break em Loops
for (i in 1:10) {
  if (i == 8) {
    break
  }
  print(i)
}
```

```r
# Next em Loops
for (i in 1:10) {
  if (i %% 2 == 0) {
    next
  }
  print(i)
}
```

```r
# Função personalizada
soma <- function(a, b) {
  return(a + b)
}

resultado <- soma(3, 5)
resultado
```

```r
# Aplicar funcao a um vetor
numeros <- c(1, 2, 3, 4, 5)

quadrado <- function(x) {
  return(x^2)
}

sapply(numeros, quadrado)
```

```r
# Condicional com operador ternario
resultado <- ifelse(5 > 3, "Verdadeiro", "Falso")
resultado
```

```r
# Criação de uma sequencia
sequencia <- seq(0, 10, by = 2)
sequencia
```

```r
# Repetição de elementos
repetidos <- rep(1:3, times=3)
repetidos
```

```r
# Acessando elementos de um vetor
vetor <- c("a", "b", "c", "d")

elemento <- vetor[2]
elemento
```

```r
vetores <- c("a", "b", "c", "d")

for (vetor in vetores) {
  print(vetor)
}
```

```r
# Acessando elementos de uma lista
minha_lista <- list(nome = "Ana", idade = 25)

idade <- minha_lista$idade
idade
```

```r
# Acessando colunas de um dataframe
df <- data.frame(nome=c("Jaqueline", "João"), idade=c("25", "32"))

df_filtrado <- subset(df, idade > 25)
df_filtrado
```

```r
# Ordenando um DataFrame
df <- data.frame(nome = c("Ana", "João"), idade=c(25, 30))

df_ordenado <- df[order(df$idade),]
df_ordenado
```

```r
# Aplicando uma funcao a colunas de um DataFrame
df <- data.frame(a=1:5, b=6:10)
colSums(df)
```

```r
# Mesclando DataFrames
df1 <- data.frame(id = 1:2, valor1 = c(10, 20))
df2 <- data.frame(id = 1:2, valor2 = c(30, 40))

df_mesclado <- merge(df1, df2, by = "id")


df_mesclado
```

```r
# Transpondo uma Matriz
matriz <- matrix(1:4, nrow=2, ncol=2)
matriz_transposta <- t(matriz)
matriz
```

```r
# Reshape de um DataFrame
install.packages("reshape2")
library(reshape2)

df <- data.frame(id=1:2, tempo1=c(30, 40), tempo2=c(50, 60))

df_melt <- melt(df, id.vars="id", variable.name="tempo", value.name="valor")
print(df_melt)
```

```r
# Renomeando uma coluna de um DataFrame
df <- data.frame(a=1:5, b=6:10)

names(df) <- c("coluna1", "coluna2")
```

```r
# Removendo colunas de um data frame
df <- data.frame(a = 1:5, b = 6:10)
df
df <- df[, -2]
df
```
